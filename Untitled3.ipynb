```python
df=pd.read_csv("openings.csv",index_col=0)
cols=["Opening","Colour","Num Games","Perf Rating","Avg Player","Player Win %","Draw %","Opponent Win %"]

df=df[cols]
df["performance"]=df["Perf Rating"]-df["Avg Player"]
df.head()
```

```python
def add(a,b):
  return a+b
```

```python
def abb(names):
  abbre=[]
  for i in names:
    i=str(i)
    words=i.split()
    fc=[re.sub(r'\W+', '', i)[0] for i in words]
    abbrevation="".join(fc)
    abbre.append(abbrevation)
  return abbre

```

```python
abb("abbrevation python")
```

```python
win_white=df.sort_values(by=["Player Win %"],ascending=False).head(n=4)[["Opening","Player Win %"]]
win_black=df.sort_values(by=["Opponent Win %"],ascending=False).head(n=4)[["Opening","Opponent Win %"]]
draw=df.sort_values(by=["Draw %"],ascending=False).head(n=4)[["Opening","Draw %"]]

#sns.set_style("darkgrid")
plt.figure(figsize=(12,6))
plt.bar(x=abb(win_white["Opening"]),height=win_white["Player Win %"],color="white",label="white wins")
plt.bar(x=abb(win_black["Opening"]),height=win_black["Opponent Win %"],color="black",label="black wins")
plt.bar(x=abb(draw["Opening"]),height=draw["Draw %"],color="red",label="draw")
plt.xlabel("Opening",fontsize=12)
plt.ylabel("Probability of winning %",fontsize=12)
plt.title("Most favourate opening by expected result")
plt.xticks(rotation=105,ha="left")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# Explain abbreviations
print("Abbreviations:")
names = win_white["Opening"].tolist() + draw["Opening"].tolist() + win_black["Opening"].tolist()
abbreviations = abb(names)
for abbreviation, name in zip(abbreviations, names):
    print(f'- {abbreviation} - {name}')
```

```python
import matplotlib.pyplot as plt

# Sample data
x = ['Opening 1', 'Opening 2', 'Opening 3']
win_white = {'Opening': x, 'Player Win %': [30, 40, 50]}
draw = {'Opening': x, 'Draw %': [10, 20, 15]}
win_black = {'Opening': x, 'Opponent Win %': [20, 30, 35]}

# Plotting
plt.figure(figsize=(12, 6))

bar_width = 0.3
index = range(len(x))

plt.bar(index, win_white['Player Win %'], bar_width, color='green', label='White win')
plt.bar(index, draw['Draw %'], bar_width, color='yellow', label='Draw', bottom=win_white['Player Win %'])
plt.bar(index, win_black['Opponent Win %'], bar_width, color='red', label='Black win',
        bottom=[win_white['Player Win %'][i] + draw['Draw %'][i] for i in range(len(x))])

plt.xlabel('Opening', fontsize=12)
plt.ylabel('Probability of winning [%]', fontsize=12)
plt.title('Most favorable openings by expected result', fontsize=16)
plt.xticks(index, x, rotation=90, ha='right')
plt.tight_layout()
plt.legend()
plt.show()
```
